````python
# ─── GLOBAL STYLE ────────────────────────────────────────────

plt.rcParams.update({
    'font.size':        22,
    'axes.titlesize':   28,
    'axes.labelsize':   26,
    'legend.fontsize':  18,
    'xtick.labelsize':  22,
    'ytick.labelsize':  22,
    'figure.dpi':       100,
    'savefig.dpi':      DPI,
    'savefig.bbox':     'tight',
    'figure.facecolor': 'white',
})

COL_RUWE = '#EE6677'
COL_DIST = '#228833'
COL_RV   = '#4477AA'
COL_COMB = '#222222'

def style_ax(ax, minor=True):
    """Apply large-font, big-tick paper style to any axis."""
    for sp in ax.spines.values():
        sp.set_linewidth(2.5)
    ax.tick_params(axis='both', which='major',
                   labelsize=22, width=2.8, length=14,
                   direction='in', top=True, right=True)
    ax.tick_params(axis='both', which='minor',
                   width=1.8,   length=8,
                   direction='in', top=True, right=True)
    if minor:
        ax.minorticks_on()
    for item in [ax.xaxis.label, ax.yaxis.label]:
        item.set_fontsize(26)
        item.set_fontweight('bold')
    if ax.get_title():
        ax.title.set_fontsize(28)
        ax.title.set_fontweight('bold')

def style_ax_inset(ax, minor=True):
    """Style for inset axes: large but slightly smaller than main."""
    for sp in ax.spines.values():
        sp.set_linewidth(2.2)
    ax.tick_params(axis='both', which='major',
                   labelsize=18, width=2.2, length=10,
                   direction='in', top=True, right=True)
    ax.tick_params(axis='both', which='minor',
                   width=1.4,   length=6,
                   direction='in', top=True, right=True)
    if minor:
        ax.minorticks_on()
    for item in [ax.xaxis.label, ax.yaxis.label]:
        item.set_fontsize(20)
        item.set_fontweight('bold')
    if ax.get_title():
        ax.title.set_fontsize(22)
        ax.title.set_fontweight('bold')

def savefig(fig, name):
    if SAVE_FIGS:
        p = os.path.join(OUTPUT_DIR, name)
        fig.savefig(p, dpi=DPI, bbox_inches='tight')
        print(f'  Saved: {p}')
    plt.show()

# ─── PDF helpers (reconstruct from saved params) ─────────────

def ln_core(x, p):   return p['eta'] * stats.lognorm.pdf(x, s=p['sigma_core'], scale=np.exp(p['mu_core']))
def ln_tail(x, p):   return (1-p['eta']) * stats.lognorm.pdf(x, s=p['sigma_tail'], scale=np.exp(p['mu_tail']))
def gm_core(x, p):   return p['eta'] * stats.gamma.pdf(x, a=p['alpha_core'], scale=1./p['beta_core'])
def gm_tail(x, p):   return (1-p['eta']) * stats.gamma.pdf(x, a=p['alpha_tail'], scale=1./p['beta_tail'])
def ga_core(x, p):   return p['eta'] * stats.norm.pdf(x, p['mu_core'], p['sigma_core'])
def ga_tail(x, p):   return (1-p['eta']) * stats.norm.pdf(x, p['mu_tail'], p['sigma_tail'])

def crossover(x, cy, ty):
    idx = np.where((cy[1:] < ty[1:]) & (cy[:-1] >= ty[:-1]))[0]
    return x[idx[0]] if len(idx) else None

def purity_gauss(x, p):
    lc = np.log(p['eta']+1e-10) - 0.5*(x-p['mu_core'])**2/max(p['sigma_core']**2,1e-10)
    lt = np.log(1-p['eta']+1e-10) - 0.5*(x-p['mu_tail'])**2/max(p['sigma_tail']**2,1e-10)
    return np.exp(lc - np.logaddexp(lc, lt))

def purity_pdf(cy, ty):  tot = cy+ty; return np.where(tot>1e-30, cy/tot, 0.5)

print('Helpers ready.')
````

## Plot 1 — Mixture Decomposition: LN (top) vs Gamma (bottom)

````python
fig, axes = plt.subplots(2, 3, figsize=(28, 18))


x_ruwe = np.linspace(0.5, 5.0, 2000)
x_err  = np.linspace(0.001, 1.5, 2000)

def plot_hist(ax, col, xlim):
    if 'metric_histograms' in EX and col in EX['metric_histograms']:
        h = EX['metric_histograms'][col]
        edges  = np.array(h['edges'])
        counts = np.array(h['counts'])
        centers = 0.5*(edges[:-1]+edges[1:])
        ax.bar(centers, counts, width=np.diff(edges), alpha=0.5,
               color='steelblue', label='Data', align='center')

def mix_panel(ax, x, core_fn, tail_fn, params, hist_col, xlim,
              xlabel, title, vline=None, vline_label=None,
              show_cross=False):
    plot_hist(ax, hist_col, xlim)
    cy = core_fn(x, params)
    ty = tail_fn(x, params)
    ax.fill_between(x, cy, alpha=0.3, color='green',
                    label=f'Core ($\\eta$={params["eta"]:.2f})')
    ax.fill_between(x, ty, alpha=0.3, color='red', label='Tail')
    ax.plot(x, cy+ty, 'k--', lw=2.5, label='Total')
    if show_cross:
        cx = crossover(x, cy, ty)
        if cx is not None:
            ax.axvline(cx, color='purple', ls='-.', lw=2.5,
                       label=f'Cross={cx:.3f}')
    if vline is not None:
        ax.axvline(vline, color='orange', ls=':', lw=2.5, label=vline_label)
    ax.set_xlabel(xlabel)
    ax.set_ylabel('Density')
    ax.set_title(title)
    ax.set_xlim(xlim)
    ax.legend(loc='upper right', fontsize=16)
    style_ax(ax)

for row in range(2):
    ax = axes[row, 0]
    if 'ruwe' in models:
        mix_panel(ax, x_ruwe, ga_core, ga_tail, models['ruwe'],
                  'ruwe', (0.5, 5.0), 'RUWE',
                  'RUWE — Reg. Gaussian',
                  vline=1.4, vline_label='RUWE=1.4')

if 'dist_ln' in models:
    mix_panel(axes[0,1], x_err, ln_core, ln_tail, models['dist_ln'],
              'rel_dist_err', (0.0, 1.5), r'$\sigma_d/d$',
              r'$\sigma_d/d$ — Log-Normal', show_cross=True)

if 'rv_ln' in models:
    mix_panel(axes[0,2], x_err, ln_core, ln_tail, models['rv_ln'],
              'rel_rv_err', (0.0, 1.5), r'$\sigma_v/v$',
              r'$\sigma_v/v$ — Log-Normal', show_cross=True)

if 'dist_gm' in models:
    mix_panel(axes[1,1], x_err, gm_core, gm_tail, models['dist_gm'],
              'rel_dist_err', (0.0, 1.5), r'$\sigma_d/d$',
              r'$\sigma_d/d$ — Gamma', show_cross=True)

if 'rv_gm' in models:
    mix_panel(axes[1,2], x_err, gm_core, gm_tail, models['rv_gm'],
              'rel_rv_err', (0.0, 1.5), r'$\sigma_v/v$',
              r'$\sigma_v/v$ — Gamma', show_cross=True)

plt.tight_layout(rect=[0, 0, 1, 0.96])
savefig(fig, 'plot1_mixture_comparison.png')
````

## Plot 2 — Purity Curves (left) + Purity Distributions (right)

````python
fig, axes = plt.subplots(1, 2, figsize=(26, 11))

# ── LEFT: P_core(x) curves ───────────────────────────────────
ax = axes[0]
x_ruwe = np.linspace(0.5, 5.0, 1000)
x_err  = np.linspace(0.001, 1.5, 1000)

if 'ruwe' in models:
    ax.plot(x_ruwe, purity_gauss(x_ruwe, models['ruwe']),
            color=COL_RUWE, lw=3.0, ls='-', label='RUWE (RegGauss)')

if 'dist_ln' in models:
    cy = ln_core(x_err, models['dist_ln']); ty = ln_tail(x_err, models['dist_ln'])
    ax.plot(x_err, purity_pdf(cy, ty), color=COL_DIST, ls='--', lw=3.0, label=r'Dist (LN)')
if 'dist_gm' in models:
    cy = gm_core(x_err, models['dist_gm']); ty = gm_tail(x_err, models['dist_gm'])
    ax.plot(x_err, purity_pdf(cy, ty), color=COL_DIST, ls='-',  lw=3.0, label=r'Dist ($\Gamma$)')

if 'rv_ln' in models:
    cy = ln_core(x_err, models['rv_ln']); ty = ln_tail(x_err, models['rv_ln'])
    ax.plot(x_err, purity_pdf(cy, ty), color=COL_RV, ls='--', lw=3.0, label=r'RV (LN)')
if 'rv_gm' in models:
    cy = gm_core(x_err, models['rv_gm']); ty = gm_tail(x_err, models['rv_gm'])
    ax.plot(x_err, purity_pdf(cy, ty), color=COL_RV, ls='-',  lw=3.0, label=r'RV ($\Gamma$)')

ax.axhline(0.5, color='gray', ls='--', alpha=0.4)
ax.set_xlabel('Error metric value')
ax.set_ylabel(r'$P_{\rm core}$ (Purity)')
ax.set_title('Purity Curves — All Metrics')
ax.set_ylim(-0.05, 1.05)
ax.legend(fontsize=16)
ax.grid(True, alpha=0.2)
style_ax(ax)

# ── RIGHT: purity distributions ──────────────────────────────
ax = axes[1]
ph = EX.get('purity_histograms', {})

col_map = [
    ('purity_ruwe',     COL_RUWE, 'RUWE',                    2.5),
    ('purity_dist',     COL_DIST, r'Distance ($\Gamma$)',    2.5),
    ('purity_rv',       COL_RV,   r'RV ($\Gamma$)',          2.5),
    ('purity_combined', COL_COMB, 'Combined',                 4.0),
]
for col, clr, lab, lw in col_map:
    if col in ph:
        h = ph[col]
        edges  = np.array(h['edges'])
        counts = np.array(h['counts'])
        mu = h['mean']
        ax.step(edges[:-1], counts/1e6, where='post', color=clr, lw=lw,
                label=f'{lab} ($\\mu$={mu:.3f})')
    elif ph:
        print(f'  {col} not in extra_plot_data.json')

ax.set_xlabel('Purity')
ax.set_ylabel(r'Count [$\times 10^6$]')
ax.set_title('Purity Distributions')
ax.set_xlim(0, 1)
ax.legend(fontsize=16)
ax.grid(True, alpha=0.2)
style_ax(ax)

plt.tight_layout()
savefig(fig, 'plot2_purity_curves_and_distributions.png')
````

## Plot 3 — Radial Profiles (dist / rv / ruwe)

````python
rep = EX.get('radial_error_profiles', {})

for key, mcol, title, xlabel in [
    ('dist', 'rel_dist_err', r'$\sigma_d/d$', 'Distance Error'),
    ('rv',   'rel_rv_err',   r'$\sigma_v/v$', 'RV Error'),
    ('ruwe', 'ruwe',         'RUWE',            'RUWE'),
]:
    rk = key + '_radial'
    gk = key + '_global'
    if rk not in tr or gk not in tr:
        print(f'Skipping {key} — tradeoff data missing'); continue

    # ── Square panels: 2 stacked square axes ─────────────────────────────
    # Each panel is ~9 × 9 inches → figure is ~9 × 20 (with spacing)
    fig = plt.figure(figsize=(8, 16), facecolor='white')
    gs  = gridspec.GridSpec(2, 1, figure=fig,
                             hspace=0.38,
                             left=0.14, right=0.88,
                             top=0.94,  bottom=0.06)
    ax_pur = fig.add_subplot(gs[0])
    ax_ret = fig.add_subplot(gs[1])

    # Force each axes box to be square
    ax_pur.set_box_aspect(1)
    ax_ret.set_box_aspect(1)

    rd       = tr[rk]; gd = tr[gk]
    rc       = np.array(rd['r_centers'])
    bc       = rd['by_cut']
    cut_keys = sorted(bc.keys(), key=float)
    clrs     = plt.cm.viridis(np.linspace(0, 0.9, len(cut_keys)))

    # ── TOP: Radial purity profile ────────────────────────────────────────
    for ck, clr in zip(cut_keys, clrs):
        pp = np.array(bc[ck]['purity']); vm = np.isfinite(pp)
        if np.sum(vm) > 3:
            ax_pur.plot(rc[vm], pp[vm]*100, '-', color=clr, lw=2.0,
                        label=f'{title}<{float(ck):.2f}')
    ax_pur.set_ylabel('Mean Purity [%]', fontsize=26, fontweight='bold')
    ax_pur.set_title(f'Radial Purity Profile — {xlabel} Cuts',
                     fontsize=28, fontweight='bold', pad=14)
    ax_pur.set_xscale('log')
    ax_pur.set_ylim(0, 105)
    ax_pur.grid(True, alpha=0.2)
    ax_pur.legend(loc='lower left', fontsize=14, ncol=3,
                  framealpha=0.95)
    style_ax(ax_pur)

    # ── INSET top panel — square, opaque, bold labels ─────────────────────
    # width/height in axis-fraction coordinates for square appearance
    ax_i1 = inset_axes(ax_pur,
                       width='36%', height='36%',
                       loc='upper right',
                       borderpad=2.2,
                       bbox_to_anchor=ax_pur.bbox,
                       bbox_transform=ax_pur.transAxes)
    # Fully opaque white background so main plot does NOT bleed through
    ax_i1.set_facecolor('white')
    ax_i1.patch.set_facecolor('white')
    ax_i1.patch.set_alpha(1.0)
    for sp in ax_i1.spines.values():
        sp.set_linewidth(2.0)
        sp.set_edgecolor('black')
        sp.set_zorder(10)

    ax_i1.plot(gd['cuts'], [r*100 for r in gd['retention']],
               'o-', color='#2196F3', lw=2.5, ms=6, label='Retention')
    ax_i1.plot(gd['cuts'],
               [p*100 if not (p is None or (isinstance(p, float) and np.isnan(p))) else 0
                for p in gd['mean_purity']],
               's--', color='#F44336', lw=2.5, ms=6, label='Purity')
    ax_i1.plot(gd['cuts'], [c*100 for c in gd['model_completeness']],
               'D:', color='#228833', lw=2.5, ms=6, label='Model Compl.')
    ax_i1.set_xlabel(f'{title} cut', fontsize=20, fontweight='bold', labelpad=5)
    ax_i1.set_ylabel('%', fontsize=20, fontweight='bold', labelpad=5)
    ax_i1.set_title('Trade-off', fontsize=22, fontweight='bold', pad=8)
    ax_i1.set_ylim(0, 105)
    ax_i1.legend(fontsize=14, loc='center right', framealpha=1.0)
    ax_i1.grid(True, alpha=0.3)
    ax_i1.set_box_aspect(1)   # ← square inset
    style_ax_inset(ax_i1)

    # ── BOTTOM: Radial retention profile ─────────────────────────────────
    for ck, clr in zip(cut_keys, clrs):
        rr = np.array(bc[ck]['retention']); vm = np.isfinite(rr)
        if np.sum(vm) > 3:
            ax_ret.plot(rc[vm], rr[vm]*100, '--', color=clr, lw=2.0,
                        label=f'{title}<{float(ck):.2f}')
    ax_ret.set_xlabel('Galactocentric Distance R [kpc]',
                       fontsize=26, fontweight='bold', labelpad=8)
    ax_ret.set_ylabel('Retention [%]', fontsize=26, fontweight='bold')
    ax_ret.set_title(f'Radial Retention Profile — {xlabel} Cuts',
                     fontsize=28, fontweight='bold', pad=14)
    ax_ret.set_xscale('log')
    ax_ret.set_ylim(0, 105)
    ax_ret.grid(True, alpha=0.2)
    ax_ret.legend(loc='lower left', fontsize=14, ncol=3,
                  framealpha=0.95)
    style_ax(ax_ret)

    # ── INSET bottom panel — square, opaque, bold labels ─────────────────
    ax_i2 = inset_axes(ax_ret,
                       width='36%', height='36%',
                       loc='upper right',
                       borderpad=2.2,
                       bbox_to_anchor=ax_ret.bbox,
                       bbox_transform=ax_ret.transAxes)
    ax_i2.set_facecolor('white')
    ax_i2.patch.set_facecolor('white')
    ax_i2.patch.set_alpha(1.0)
    for sp in ax_i2.spines.values():
        sp.set_linewidth(2.0)
        sp.set_edgecolor('black')
        sp.set_zorder(10)

    if mcol in rep:
        rp   = rep[mcol]
        rcm2 = np.array(rp['r_centers'], dtype=float)
        meds = np.array([v if v is not None else np.nan for v in rp['median']])
        p16s = np.array([v if v is not None else np.nan for v in rp['p16']])
        p84s = np.array([v if v is not None else np.nan for v in rp['p84']])
        vm2  = np.isfinite(meds)
        if np.sum(vm2) > 2:
            ax_i2.plot(rcm2[vm2], meds[vm2], 'b-', lw=2.5, label='Median')
            ax_i2.fill_between(rcm2[vm2], p16s[vm2], p84s[vm2],
                               alpha=0.25, color='blue')
    ax_i2.set_xscale('log')
    ax_i2.set_xlabel('R [kpc]', fontsize=20, fontweight='bold', labelpad=5)
    ax_i2.set_ylabel(f'Median {title}', fontsize=20, fontweight='bold', labelpad=5)
    ax_i2.set_title('Error Profile', fontsize=22, fontweight='bold', pad=8)
    ax_i2.set_box_aspect(1)   # ← square inset
    ax_i2.grid(True, alpha=0.3)
    ax_i2.legend(fontsize=14, loc='upper left', framealpha=1.0)
    style_ax_inset(ax_i2)

    plt.savefig(os.path.join(OUTPUT_DIR, f'plot3_radial_{key}.png'),
                dpi=72, bbox_inches='tight')
    print(f'  Saved: plot3_radial_{key}.png')
    # plt.show()
````
